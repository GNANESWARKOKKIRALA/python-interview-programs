```python
"""1️⃣ Stack Problem – Push and Pop Elements
Problem

Create a stack, add 3 numbers, and remove the top element.

Explanation

Push → add element to stack

Pop → remove last element

Stack follows LIFO (Last In First Out).

Example:
Push → 10, 20, 30
Stack → [10, 20, 30]
Pop → removes 30"""
stack=[]
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)
stack.pop(-1)
print(stack)
print(stack[-1])
```

```python
"""3️⃣ Queue Problem – Enqueue and Dequeue
Problem

Create a queue, add 3 numbers, then remove the first element.

Explanation

Queue follows FIFO (First In First Out).

Example
Add → 10, 20, 30
Queue → [10, 20, 30]

Dequeue → removes 10"""
from collections import deque
queue=deque()
queue.append(10)
queue.append(20)
queue.append(30)
print(queue)
queue.popleft()
print(queue)
print(queue[0])
```

```python
"""1️⃣ Stack Problem – Reverse a String
Problem

Reverse a string using a stack.

Example
Input: "GYM"
Output: "MYG"

Explanation

Stack follows LIFO.
So the last character inserted will come out first, which reverses the string.

Steps:

Push all characters into stack

Pop characters one by one

Join them"""
stack = []
string = "GYM"

# push characters
for ch in string:
    stack.append(ch)

reversed_string = ""

# pop characters
while stack:
    reversed_string += stack.pop()

print("Reversed string:", reversed_string)
```

```python
"""📌 Problem Statement

You are given a queue of numbers.

👉 Check whether a given number is present in the queue or not.

🧾 Example

Queue = [10, 20, 30, 40]                      
Search = 30

Output:
👉 Found"""
from collections import deque
queue=deque([10,20,30,40,50])
search=40
found=False
for i in queue:
    if i==search:
        found=True
        break
if found:
    print("Found")
else:
    print("not found")
```

```python
"""📌 Problem Statement

You are given a queue of numbers.

👉 Print only even numbers from the queue.

🧾 Example

Queue = [10, 15, 20, 25, 30]"""
from collections import deque
queue=deque([10,15,20,25,30])
for i in queue:
    if i%2==0:
        print(i)
```

```python
"""📌 Problem Statement

You are given a queue of numbers.

👉 Find the sum of all elements in the queue.

🧾 Example

Queue = [10, 20, 30]"""
from collections import deque
queue=deque([10,20,30])
sum=0
for num in queue:
    sum+=num
print(sum)
```
