```python
#Armstrong number:sum of the cubes is equal to given number
num = int(input("Enter a number: "))
sum = 0  #sum will store the cube total
temp = num #temp is copy of num
while temp>0:
    digit=temp%10 # import the last digit in value
    sum+=digit**3 
    temp//=10
if sum == num:
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")
```

```python
#palindrome
num=int(input("enter the palindrome number"))
original=num
reverse=0
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num//=10
if original==reverse:
    print("the given num is an palindrome")
else:
    print("the given num is not an palinrome")
```

```python
# Factorial using Recrusive
def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
      
print(factorial(4))
```

```python
#Multiple inheritance: one child class with multiple parent class
class Father:
    def farm(self):
        print(" Father Knows farming ")
class Mother:
    def cook(self):
        print(" MOther Knows cooking")
class Child(Father,Mother):
    def play(self):
        print(" Child loves playing")
d=Child()
d.farm()
d.cook()
d.play()

```

```python
#multi level inheritance: it means a class inherits from a class, which in turn inherits from another class.
class GrandFather:
    def Rich(self):
        print("Grandfather is rich")
class Father(GrandFather):
    def Middleclass(self):
        print("Father is Middle class ")
class Son(Father):
    def poor(self):
        print("son is poor")
s= Son()
s.Rich()
s.Middleclass()
s.poor()
```

```python
#Hierarchical Inheritance: multiple child class with an parnet class
class Parent:
    def work(self):
        print("father works daily")
class Child1(Parent):
    def Gym(self):
        print("Child1 do gym")
class Child2(Parent):
    def Dance(self):
        print("Child2 do Dance")
c1=Child1()
c2=Child2()
c1.work()
c1.Gym()
c2.work()
c2.Dance()
```

```python
#Hybrid Inheritance:It is a combination of multiple types of inheritance.
class A:
    def methodA(self):
        print("Class A")

class B(A):
    def methodB(self):
        print("Class B")

class C:
    def methodC(self):
        print("Class C")

class D(B, C):
    def methodD(self):
        print("Class D")

d = D()
d.methodA()
d.methodB()
d.methodC()
d.methodD()

```

```python
#Abstraction :means hiding the internal details and showing only the essential features of an object.
#In Python, abstraction is done using abstract classes and abstract methods from the abc module.
class Car:
    def mileage(self):
        pass  # Method to be defined in child classes

class Tesla(Car):
    def mileage(self):
        print("Tesla mileage is 15 km/l")

class Suzuki(Car):
    def mileage(self):
        print("Suzuki mileage is 20 km/l")

class Duster(Car):
    def mileage(self):
        print("Duster mileage is 25 km/l")

# Create objects
t = Tesla()
s = Suzuki()
d = Duster()

# Call methods
t.mileage()
s.mileage()
d.mileage()



```

```python
""" Generator: A generator is a special type of function in Python.
In a generator, we use the yield keyword instead of return.

Yield gives values one by one, while return gives all values at once.

Using yield saves memory because it produces values on demand."""""

def my_generator(n):
    for i in range(n):
        yield i
gen=my_generator(5)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#Decorators in Python
#A decorator is a function that modifies or extends the behavior of another function without changing its code.
def decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@decorator
def hello():
    print("Hello!")

hello()

```
